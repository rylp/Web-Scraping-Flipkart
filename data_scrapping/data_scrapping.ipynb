```python
#REFERENCE :https://www.youtube.com/watch?v=mKxFfjNyj3c [EduReka Video]
#OBJECTIVE: Scrape data from a website and create an 'csv' file anstore it into it

#STEP1: IMPORTING NECCESSARY LIBRARIES

import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

import bs4  #Beautiful Soup
import urllib.request #URL Library
```

```python
from bs4 import BeautifulSoup as soup
#from urllib.request import urlopen as Ureq
```

```python
#STEP 2: CREATE URL AND READ IT IN 'source'

url='https://www.flipkart.com/search?q=iphone&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_0_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_0_6_na_na_na&as-pos=0&as-type=RECENT&suggestionId=iphone%7CMobiles&requestId=a0ad030b-a69b-4566-a24a-15cf9c9651dc&as-searchtext=iphone'
#The above is the link to flipkart with 'Iphone'searched
webpage=urllib.request.urlopen(url)
print(webpage)
#This only print the specification & not any contents
source=webpage.read()
webpage.close()
print(source)
#This prints contents but not in a view-able order
```

```python
#STEP3: PARSING THE 'source' AND CREATING A 'page_soup' BY 'html.parser'
page_soup=soup(source, "html.parser")
print(page_soup)
#this print the proper html file of the webpage
```

```python
#STEP 4

all_containers=page_soup.findAll("div",{"class": "_3O0U0u"})
print(all_containers)
#This displays all the stuff we've just scrapped

#Now there must be 24 entries so check length

print(len(all_containers))
#So it prints 24 at the end
```

```python
#STEP 5

#take the first of 'all_containers'and put it into a seperate var to scrap

print(soup.prettify(all_containers[0]))
#so,We get the good readable tml type file for the first container

one_container=all_containers[0]
#storing 1st container from 'all_containers' to 'one_container' var
#print(one_container)
```
