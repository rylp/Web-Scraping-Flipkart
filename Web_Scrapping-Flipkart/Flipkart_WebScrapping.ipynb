```python
#OBJECTIVE: Scrape data from a website and create an 'csv' file anstore it into it
#STEP1: IMPORTING NECCESSARY LIBRARIES

#Iphones on flipkart
#Site: https://www.flipkart.com/search?q=iphone&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_0_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_0_6_na_na_na&as-pos=0&as-type=HISTORY&suggestionId=iphone%7CMobiles&requestId=a0ad030b-a69b-4566-a24a-15cf9c9651dc&as-backfill=on
#MAKE A DATA FILE IN WHICH WE HAVE iPHONE model name , RATING , PRICE

import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

import bs4  #Beautiful Soup
import urllib.request #URL Library
```

```python
from bs4 import BeautifulSoup as soup
#from urllib.request import urlopen as Ureq
```

```python
#STEP 2: CREATE URL AND READ IT IN 'source'

url='https://www.flipkart.com/search?q=iphone&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_0_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_0_6_na_na_na&as-pos=0&as-type=RECENT&suggestionId=iphone%7CMobiles&requestId=a0ad030b-a69b-4566-a24a-15cf9c9651dc&as-searchtext=iphone'
#The above is the link to flipkart with 'Iphone'searched
webpage=urllib.request.urlopen(url)
print(webpage)
#This only print the specification & not any contents
source=webpage.read()
webpage.close()
print(source)
#This prints contents but not in a view-able order
```

```python
#STEP3: PARSING THE 'source' AND CREATING A 'page_soup' BY 'html.parser'
page_soup=soup(source, "html.parser")
print(page_soup)
#this print the proper html file of the webpage
```

```python
#STEP 4 TAKE THE REQUIRED CONTENT INTO A VAR BY USING APPROPRIATE html tags

all_containers=page_soup.findAll("div",{"class": "_3O0U0u"}) #here the Data Required is in div with class:_3O0U0u
print(all_containers)
#This displays all the stuff we've just scrapped

#Now there must be 24 entries so check length

print(len(all_containers))
#So it prints 24 at the end
```

```python
#STEP 5

#take the first of 'all_containers'and put it into a seperate var to scrap

print(soup.prettify(all_containers[0]))
#so,We get the good readable tml type file for the first container

one_container=all_containers[0]
#storing 1st container from 'all_containers' to 'one_container' var
#print(one_container)
```

```python
#STEP 6

product=one_container.div.img["alt"]
print(product)
#So we got the tile

#we didnt have to use a findAll in this as the only <div> tag with an <img alt="Phone Model"> was present in the text file

#Now going for rating
```

```python
#We will have to use findAll in this case

ratings=one_container.findAll("div",{"class":"niH0FQ"}) #So, search for that 'div' and 'class:niH0FQ' in the file .
#As we have only 1 occurence hence ratings[0] will have the below '4.583,921 Ratings & 8,548 Reviews' data
#ratings[1] and so on dont exist
#but still have to print it using array indexing ie ratings[0]
print(ratings[0].text)

#now go for price
```

```python
price=one_container.findAll("div",{"class":"_6BWGkk"})
print(price[0].text)
#Printed the price
```

```python
#STEP 7: CREATE A .csv file and WRITE THIS DATA OBTAINED IN THIS FILE WITH PROPER HEADERS

filename="products.csv"
f=open(filename,"w")

headers="Product,Pricing,Ratings\n"
f.write(headers)
```

```python
#STEP 8: DO IT FOR all_containers

#now we want to do this for all 24 of the iPhones
#so apply for loop in all_containers DataFrame

for one_container in all_containers:
    
    container_product=one_container.div.img["alt"]
    product=container_product
    #print("Product=> "+product)
    
    container_price=one_container.findAll("div",{"class":"_6BWGkk"})
    price=container_price[0].text.strip() #strip() is use dto eliminate unnecessary commas and punctuation
    #print(price)
    
    container_ratings=one_container.findAll("div",{"class":"niH0FQ"})
    ratings=container_ratings[0].text
    #print(ratings)
    
    #Lets go for making the price display better like=>'Rs.27500'
    
    add_rs = "Rs." + price #Adding Rs. to each 
    split_by_rs=add_rs.split("") #Splitting by '' symbol. So split_by_rs[0]='Rs .' and split_by_rs[1]='27,999'
    #print("Pricing=> "+split_by_rs[0]+split_by_rs[1]) #So, combine [0] & [1]
    
    #SIDENOTE => CTRL+ALT+4 => ''
    
    #Lets go for making the ratings display better like=> '4.583'
    split_by_comma=ratings.split(",")
    #print("Ratings=> "+split_by_comma[0])
    
    #Now,replace commas(,) with tilde(|)
    print(product.replace(",","|") + "," + split_by_rs[0] + split_by_rs[1].replace("," ,"") + "," + split_by_comma[0]+"\n")
    f.write(product.replace(",","|") + "," + split_by_rs[0] + split_by_rs[1].replace(",","") + "," + split_by_comma[0]+"\n")
```

```python
f.close()
```
